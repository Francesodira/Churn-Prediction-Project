Udersampling

```python
import pandas as pd
from sklearn.utils import resample
majority_class = df[df['Exited'] == 0]
minority_class = df[df['Exited'] == 1]
# Determine the number of instances in the minority class
minority_count = len(minority_class)
# Undersample the majority class to match the minority class
majority_undersampled = resample(majority_class, replace=False, n_samples=minority_count, random_state=42)

# Combine the minority class and the undersampled majority class into a balanced dataset
under_df = pd.concat([majority_undersampled, minority_class])

# Now 'balanced_df' contains your balanced dataset
```

```python
u_x = under_df.drop(['Exited','HasCrCard','EstimatedSalary'], axis=1)

u_y = under_df.Exited
```
